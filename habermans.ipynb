# CDF

```python
#CDF for class 1.
y = x.loc[x["surv_status"] == 1]
z = x.loc[x["surv_status"] == 2]

counts, bin_edges = np.histogram(y['age'], bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label = 'cdf')
plt.plot(bin_edges[1:],pdf,label = 'pdf')
plt.xlabel('Age')
plt.ylabel("percentage")
plt.title("age feature corresponding to class 1")
plt.legend()
plt.show()
```

### conclusion:
    1. it is clear that patients having age between 50 to 60 have greater rate of survival.
    2. almost 90% of cancer patients survival have age less than or eual to 70.

```python
#CDF for class 2.
y = x.loc[x["surv_status"] == 1]
z = x.loc[x["surv_status"] == 2]

counts, bin_edges = np.histogram(z['age'], bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label = 'cdf')
plt.plot(bin_edges[1:],pdf,label = 'pdf')
plt.xlabel('Age')
plt.ylabel("percentage")
plt.title("age feature corresponding to class 2")
plt.legend()
plt.show()
```

### observations:
it is cleared that less than or equal to 74% cancer patients died at age below 60years.
95% patients dies at age near 70 years.
more than 75 years 100% patients dies due to cancer.

# MEAN, VARIANCE & STANDARD DEVIATIONS

```python
print("Means: ")
print(np.mean(x["age"]))
print(np.mean(x["years"]))
print(np.mean(x["axil_node"]))
print(np.mean(np.append(x["axil_node"],25)))


print("\nStandard deviation: ")   #std dev of a square root the average distance of all the points from the mean.
print(np.std(x["age"]))
print(np.std(x["years"]))
print(np.std(x["axil_node"]))
```

# Median,Percentile,Quantile,IQR,MAD

```python
print("\nMedians:")
print(np.median(x["age"]))
print(np.median(np.append(x["age"],60)))
print(np.median(x["years"]))
print(np.median(x["axil_node"]))

print("\nQuantile:")
print(np.percentile(x['age'],np.arange(0,101,25)))
print(np.percentile(x['years'],np.arange(0,101,25)))
print(np.percentile(x['axil_node'],np.arange(0,101,25)))
print(np.percentile(x['surv_status'],np.arange(0,101,25)))

print("\nDifferent percentile:")
print(np.percentile(x["age"],90))
print(np.percentile(x["axil_node"],90))
print(np.percentile(x["years"],80))
print(np.percentile(x["surv_status"],90))

from statsmodels import robust
print("\nMedian Absolute")
print(robust.mad(x["age"]))
print(robust.mad(x["years"]))
print(robust.mad(x["axil_node"]))




```
